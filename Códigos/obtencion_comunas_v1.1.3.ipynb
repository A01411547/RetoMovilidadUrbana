```python
import numpy as np
import pandas as pd
import os
import geocoder
from geopy.geocoders import Nominatim
from geopy.point import Point
```

```python
geolocator = Nominatim(user_agent="test")

def reverse_geocoding(lat, lon):
    try:
        location = geolocator.reverse(Point(lat, lon))
        w =location.raw['display_name']
        return w
    except:
        return None
```
