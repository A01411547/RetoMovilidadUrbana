```python
import pandas as pd
```

```python
df = pd.read_csv("matriz_viajes_definitiva.csv")
```

```python
df = df[["hora_inicio","hora_fin","distancia","dir_o","dir_d"]]
df
```

```python
df["distancia"]=df["distancia"].astype(int)
```

```python
df = df[df['distancia']<25.0]
df = df[df['distancia']>0.0]
df = df[df['dir_d'] != df['dir_d'].iloc[1936390]]
df = df[(df['dir_o'] != df['dir_d'].iloc[1936390])]
df
```

```python
df = df.reset_index(drop=True)
```

```python
df.to_csv("matriz_viaje_final_final.csv")
```
