```python
import pandas as pd
```

```python
df = pd.read_csv("viajes_distintas_comunas_dp3.csv")
```

```python
df = df.rename(columns={"0":"Num_viajes"})
df = df[["dir_d", "Hora_llegada", "Num_viajes"]]
df
```

```python
df = df.set_index("dir_d", "Hora_llegada")
```

```python
df = df.sort_values(by=["dir_d","Hora_llegada"])
```

```python
df
```

```python
df2 = df.groupby(["dir_d","Hora_llegada"]).sum()
```

```python
df2
```

```python
#df2.to_csv("suma_viajes.csv")
```

```python
df2= df2.reset_index()
```

```python
dum = pd.get_dummies(df2, columns=['Hora_llegada'], drop_first=True)
```

```python
dum
```

```python
#dum.to_csv("dummies.csv")
```

```python
df2 = pd.read_csv("Posibles_destinos.csv")
```

```python
df2 = df2.sort_values(by=["Comuna"])
```

```python
df2.rename(columns = {'Comuna':'dir_d'}, inplace = True)

df2.head()

df2 = df2.drop(columns="Unnamed: 0")
```

```python
df2["dir_d"]=df2['dir_d'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
```

```python
df3 = pd.merge(dum,df2,on=["dir_d"])
df3
```

```python
df3['Num_escuelas'] = df3['Num_escuelas'].fillna(0)
df3['Num_hospitales'] = df3['Num_hospitales'].fillna(0)
df3['Num_iglesias'] = df3['Num_iglesias'].fillna(0)
df3['Num_zonas_tipicas'] = df3['Num_zonas_tipicas'].fillna(0)
df3
```

```python
df3.to_csv("training_dataset_3_new.csv", index = False)
```
